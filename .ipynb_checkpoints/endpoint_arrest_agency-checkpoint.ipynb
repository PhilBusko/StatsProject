## Import Lookups

```python
OFFENSE_MEDIUM = 'data/lookup_offense_medium.csv'
STATE = 'data/lookup_state.csv'
AGENCY = 'data/lookup_agency.csv'
YEAR_SINCE = '2018'
YEAR_UNTIL = '2018'
```

```python
agency_df = pd.read_csv(AGENCY)
agency_df.head()
agency_df.info()
```

```python
agency_subset_df = agency_df.loc[agency_df['state_abbr'].isin(['CT'])]
agency_subset_df.head()
agency_subset_df.info()
```

```python
state_df = pd.read_csv(STATE)
state_df.head()
#state_df.info()
```

```python
offense_df = pd.read_csv(OFFENSE_MEDIUM)
offense_df.head()
```

```python
offense_subset_df = offense_df.loc[offense_df['offense_type'].str.contains('drug')]
offense_subset_df
```

## Get API Data

```python
# examine data for a random row

agency_test = agency_df.loc[7410]['ori']
#state_test = state_df.loc[25]['state_abbr']
offense_test = offense_df.loc[6]['offense_type']
variable_test = 'offense'

agency_test
#state_test
offense_test
```

```python
config = {
    '{ori}': agency_test,
    '{offense}': offense_test,
    '{variable}': variable_test,   
    '{since}': YEAR_SINCE,
    '{until}': YEAR_UNTIL,
}
endpoint_test = ch.get_json(ch.ARREST_TKM_AGENCY_OFFENSE, config)
endpoint_test
```

```python
# loop over all lookup values and aggregate full results

results_ls = []
variable_type = 'offense'

for idx_agn, row_agn in agency_subset_df.iterrows():
    ori = row_agn['ori']
    print(ori)
    #break
    
    for idx_off, row_off in offense_subset_df.iterrows():
        offense = row_off['offense_type']
        
        # get the counts for males
        
        config = {
            '{ori}': ori,
            '{offense}': offense,
            '{variable}': variable_type,   
            '{since}': YEAR_SINCE,
            '{until}': YEAR_UNTIL,
        }
        result = ch.get_json(ch.ARREST_TKM_AGENCY_OFFENSE, config)
        
        if 'data' in result:       
            for res in result['data']:
                new_dx = {
                    'agency': ori,
                    'offense': offense,
                    'year': res['data_year'], 
                    'age': res['key'],
                    'count': res['value'],
                }
                results_ls.append(new_dx)
        
        else:
            print(f'NO DATA: {state} {offense} male')
        
        break

results_ls[:5]
```

## Create DataFrame from API

```python
results_ls
```

```python
endpoint_df = pd.DataFrame(results_ls)
endpoint_df = endpoint_df.sort_values(['state', 'year', 'offense', 'gender'])
endpoint_df.head(10)
endpoint_df.info()
```

```python
# no duplicates in this endpoint

doop_df = endpoint_df.loc[endpoint_df.duplicated(subset=['state', 'year', 'offense', 'gender', 'age'], keep=False) == True]
doop_df.head()
```

```python
# group the different ages into one count

endpoint_age_df = endpoint_df.groupby(['state', 'year', 'gender', 'offense']).sum()
endpoint_age_df = endpoint_age_df.reset_index()
endpoint_age_df.head(30)
```

## Export to csv

```python
save_path = r'data/arrest_tkm_agency.csv'
endpoint_age_df.to_csv(save_path, index=False)
f'{round(os.path.getsize(save_path) /1e6, 2)} mb'
```
